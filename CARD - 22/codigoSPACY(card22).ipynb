### Usando comandos da spaCy em portugues 

```python
import spacy 

NLP = spacy.load('pt_core_news_sm') # carregandoo modelo pré treinado em portugues 
```

```python
texto = 'o desempenho máximo exige mais do que condições perfeitas e sorte. A vontade de tornar-se mais rápido e forte a cada sessão de treino é decisiva. Questionando tudo, especialmente a si mesmo. O maior desafio é permanecer imbatível. Com essa atitude, a Porsche continua ultrapassando os limites do que é possível. Para aumentar ainda mais o desempenho nos circuitos - graças à incrível downforce e à leveza de toda a estrutura. Experimente o novo 911 GT3 RS em sua melhor forma.'
```

```python
print(texto)
```

```python
doc = NLP(texto)
```

```python
print(len(texto)) # tamanho do texto do arqv
print(len(doc)) # tamanho do texto no doc container 
```

```python
sentenca = list(doc.sents)[7] # convertendo para lista 
sentenca
```

```python
# retirando um token da sentença 
token = sentenca[3]
token
```

```python
# aplicando alguns atributos de token 
print(token.text)
print(token.left_edge)
print(token.ent_type_)
print(token.morph)
```

```python
# visulaizando a estrutura gramatical do texto 
from spacy import displacy
displacy.render(doc, style = 'dep')
```

```python
# visualizando as entidades
displacy.render(doc, style = 'ent')
```

```python
# usando o matcher  para encotrar Porsche no texto 
from spacy.matcher import Matcher
matcher = Matcher(NLP.vocab) # Cria um objeto Matcher e associa ele ao vocabulário do modelo
pattern = [{'LOWER': 'porsche'}] # definindo padrão 
matcher.add('PORSCHE_PATTERN', [pattern])


matches = matcher(doc)

print(len(matches)) # quantidades de correspondencias 
for match_id, inicio, fim in matches:
    span = doc[inicio:fim]
    print(f'Encontrado: {span.text}\n indices: {inicio}-{fim}\n')
```

### teste de similaridade 

```python
doc1 = NLP("Eu quero uma porsche 911 Turbo s")
doc2 = NLP("A porsche é a melhor marca de esportivo")

print(doc1, "<-->", doc2, "\nSimilaridade: ", doc1.similarity(doc2))
```

```python
doc3 = NLP("A porsche 911 Turbo possui a maior potência dos carros da porsche")

print(doc1, "<-->", doc3, "\nSimilaridade: ", doc1.similarity(doc3))
```

```python
doc4 = NLP("Eu não como maçã")
doc5 = NLP("Eu não como melancia")

print(doc4, "<-->", doc5, "\nSimilaridade: ", doc4.similarity(doc5))
```
