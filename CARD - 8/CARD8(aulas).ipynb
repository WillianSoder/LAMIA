# Web Scraping 

```python
#importando biblioteca
from bs4 import BeautifulSoup
import requests
```

```python
# Faz uma requisição à URL e obtém o conteúdo da página em formato de texto
html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords=python&txtLocation=').text 
soup = BeautifulSoup(html_text, 'lxml') # Analisa o HTML da página usando BeautifulSoup e o parser 'lxml'
```

```python
# encontrando os anuncios de emprego
jobs = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx' )
for job in jobs: # pega as informacoes de todos empregos da pagina
    company_name = job.find('h3', class_ = 'joblist-comp-name').text.replace(' ','') # encontrando a empresa do anuncio
    skills = job.find('span', class_ = 'srp-skills').text.replace(' ','') # encontrando as habilidades necessarias para o emprego 
    date = job.find('span', class_ = 'sim-posted').span.text # pegando a data de postagem do anuncio
    print(f'''
    Company name: {company_name}
    Required Skills: {skills}
    Post date: {date}
    \n\n
    ''')
```

```python
# encontrando os anuncios de emprego e filtrando 
jobs = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx' )
for job in jobs: # pega as informacoes de todos empregos da pagina
    
    date = job.find('span', class_ = 'sim-posted').span.text # pegando a data de postagem do anuncio
    
    if 'few' in date: # filtrando anuncios postados a poucos dias 
        company_name = job.find('h3', class_ = 'joblist-comp-name').text.replace(' ','') # encontrando a empresa do anuncio
        skills = job.find('span', class_ = 'srp-skills').text.replace(' ','') # encontrando as habilidades necessarias para o emprego 

        print(f'''
        Company name: {company_name}
        Required Skills: {skills}
        Post date: {date}
        \n\n
        ''')
```

```python
# encontrando os anuncios de emprego e filtrando (COM LINK)
print('Put some skill that you are not familiar with') # filtrando por habilidades
unfamiliar_skill = input('>')
print(f"Filtering out {unfamiliar_skill}")
jobs = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx' )
for job in jobs: # pega as informacoes de todos empregos da pagina
    
    date = job.find('span', class_ = 'sim-posted').span.text # pegando a data de postagem do anuncio
    
    if 'few' in date: # filtrando anuncios postados a poucos dias 
        link = job.header.h2.a['href'] # pegando o link do anuncio
        company_name = job.find('h3', class_ = 'joblist-comp-name').text.replace(' ','') # encontrando a empresa do anuncio
        skills = job.find('span', class_ = 'srp-skills').text.replace(' ','') # encontrando as habilidades necessarias para o emprego 
        
        if unfamiliar_skill not in skills:
            print(f"Company name: {company_name.strip()}")
            print(f"Required skills: {skills.strip()}")
            print(f"Post date: {date}")
            print(f"Job link: {link}")
            print('')
```

Salvando em um arquivo 

```python
#importando biblioteca
from bs4 import BeautifulSoup
import requests
import time
# encontrando os anuncios de emprego e filtrando (COM LINK)
print('Put some skill that you are not familiar with') # filtrando por habilidades
unfamiliar_skill = input('>')
print(f"Filtering out {unfamiliar_skill}")

def find_jobs():
    # Faz uma requisição à URL e obtém o conteúdo da página em formato de texto
    html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords=python&txtLocation=').text 
    soup = BeautifulSoup(html_text, 'lxml') # Analisa o HTML da página usando BeautifulSoup e o parser 'lxml'
    jobs = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx' )
    for index, job in enumerate(jobs):  # pega as informacoes de todos empregos da pagina
        
        date = job.find('span', class_ = 'sim-posted').span.text # pegando a data de postagem do anuncio
        
        if 'few' in date: # filtrando anuncios postados a poucos dias 
            link = job.header.h2.a['href'] # pegando o link do anuncio
            company_name = job.find('h3', class_ = 'joblist-comp-name').text.replace(' ','') # encontrando a empresa do anuncio
            skills = job.find('span', class_ = 'srp-skills').text.replace(' ','') # encontrando as habilidades necessarias para o emprego 
            
            if unfamiliar_skill not in skills:
                with open(f'{index}.txt', 'w') as f:
                    f.write(f"Company name: {company_name.strip()}\n")
                    f.write(f"Required skills: {skills.strip()}\n")
                    f.write(f"Post date: {date}\n")
                    f.write(f"Job link: {link}\n")
                print(f'File saved: {index}')


if __name__ == '__main__':
        find_jobs()
```
