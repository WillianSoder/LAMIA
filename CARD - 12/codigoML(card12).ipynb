```python
from sklearn.datasets import load_wine
import pandas as pd 


# extraindo o data set da biblioteca sk learn
vinhos_dataset = load_wine()
df = pd.DataFrame(data=vinhos_dataset.data, columns=vinhos_dataset.feature_names)
df['classe'] = vinhos_dataset.target # adicioando a coluna de alvos ou classes 
df
```

```python
# criando os previsores e as classes 
previsores = df.iloc[:, 0:13].values # caracteristicas 
classe = df.iloc[:, 13].values # classe do vinho (0, 1 ou 2)
```

```python
from sklearn.model_selection import train_test_split
# criando base de dados de treinamento e teste (25% da base para testar)
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe, test_size=0.25)
```

```python
# realizando o preprocessamento do dados 
from sklearn.preprocessing import StandardScaler
from keras.utils import to_categorical

# Normalizar os dados
scaler = StandardScaler()
previsores_treinamento = scaler.fit_transform(previsores_treinamento)
previsores_teste = scaler.transform(previsores_teste)

# Convertendo as classes para que as classes cotenham uma sequencia de 0's e 1 (assim como no exemplo das iris), one hot encoder
# 100
# 010
# 001
classe_treinamento = to_categorical(classe_treinamento)
classe_teste = to_categorical(classe_teste)
```

```python
from keras.models import Sequential
from keras.layers import Dense
# construindo a rede 
classificador = Sequential()
classificador.add(Dense(units = 13, activation = 'relu', input_dim = 13)) # primeira camada oculta
classificador.add(Dense(units = 13, activation = 'relu')) # segunda camada oculta
classificador.add(Dense(units = 3, activation = 'softmax')) # camada de saida
# compilando a rede 
classificador.compile(optimizer = 'adam', loss = 'categorical_crossentropy', 
                      metrics = ['categorical_accuracy'])
```

```python
# treinando a rede 
classificador.fit(previsores_treinamento, classe_treinamento, batch_size = 10, 
                  epochs = 100)
```

```python
# avaliando os resultados da rede pelo keras
resultado = classificador.evaluate(previsores_teste, classe_teste)
resultado
```

```python
import numpy as np 
previsoes = classificador.predict(previsores_teste) # fazendo o teste sem ser pelo keras
previsoes = (previsoes>0.5)# valor true e false
print(f"previsoes: {previsoes}") # exibindo com valoes true e false 

# convertendo o dados para criar a matriz de confusão 
classe_teste2 = [np.argmax(t) for t in classe_teste]
previsoes2 = [np.argmax(t) for t in previsoes]
```

```python
# contruindo a matriz de confusao para avaliar os erros e acertos da rede da rede 
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(previsoes2, classe_teste2)
print(f"Matriz de confusão\n{matriz}")
```
