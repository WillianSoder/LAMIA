### Função de ativação 

```python
# step (função degrau)
import numpy as np

# ja vai recer a soma pronta e apenas verifica 
def stepfunction(soma):
    if (soma >= 1):
        return 1
    return 0

teste = stepfunction(1)
print(teste)
teste= stepfunction(-1)
print(teste)
```

```python
# função sigmoide

def sigmoidFuncion(soma):
    return 1 / (1 + np.exp(-soma)) # usando a função de exponecial da numpy 

teste = sigmoidFuncion(1)
print(teste)
teste = sigmoidFuncion(-1)
print(teste)
```

```python
# função tangente hiperbolica 

def tahnFunction(soma):
    return (np.exp(soma) - np.exp(-soma)) / (np.exp(soma) + np.exp(-soma)) # seguindo a formula do slide

teste = tahnFunction(0.358)
print(teste)
teste = tahnFunction(-0.358)
print(teste) # retorna valores negativos 
```

```python
# função ReLU (rectified linear units) usada em CNN

def reluFunction(soma):
    if (soma >= 0):
        return soma
    return 0

teste = reluFunction(5)
print(teste)
teste = reluFunction(-5)
print(teste) 
```

```python
# função linear 
def linearFunction(soma):
    return soma

teste = linearFunction(10)
print(teste)
teste = linearFunction(-10)
print(teste)
```

```python
# função softmax (retorna probabilidade)

def softmaxFunction(x):
    ex = np.exp(x) 
    return ex / ex.sum()

valores = [5.0, 2.0, 1.3]
print(softmaxFunction(valores))
```

### Base de dados breast cancer 

```python
import pandas as pd 
import keras 
from keras.models import Sequential # classe em que os modelos tem camada de entrada, camada(s) oculta(s) e camada de saida
from keras.layers import Dense # camadas densas na rede neural
from sklearn.model_selection import train_test_split

# carregando os dados
previsores = pd.read_csv('entradas_breast.csv')
classe = pd.read_csv('saidas_breast.csv')
```

```python
# criando dados de treino e teste (25% teste)
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe, test_size=0.25)
```

```python
classificador = Sequential() # criando rede neural
# numero de neuronios (units) = (numero de entradas (num de colunas no caso 30) + numero de neuronios de saida (no caso é binario) ) / 2
classificador.add(Dense(units = 16, activation = 'relu', 
                        kernel_initializer = 'random_uniform', input_dim = 30)) # criando a primeira camada oculta 
# camada de entrada = input_dim
classificador.add(Dense(units = 1, activation = 'sigmoid')) # camada de saida 
```

```python
classificador.compile(optimizer = 'adam', loss = 'binary_crossentropy', 
                      metrics = ['binary_accuracy']) # compila a rede neural 

classificador.fit(previsores_treinamento, classe_treinamento, 
                  batch_size = 10, epochs = 100) # treinamento da rede
```

```python
previsoes = classificador.predict(previsores_teste) # realizando previsoes/teste com a rede neural 
previsoes = (previsoes > 0.5) # para fazer previsoes virar um vetor de true e false
```

```python
# medindo a precisão do modelo 
from sklearn.metrics import confusion_matrix, accuracy_score

precisao = accuracy_score(classe_teste, previsoes) # ve a precisão de previsoes em relação a classe_teste (valor real)
print(f"Acerto de {precisao*100: .5f}%")

matriz = confusion_matrix(classe_teste, previsoes) # matriz de acertos e erros de previsão em relação ao valor real 
print(matriz)
```

### Validação cruzada com dropout 

```python
import pandas as pd 
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score

# carregando os dados
previsores = pd.read_csv('entradas_breast.csv')
classe = pd.read_csv('saidas_breast.csv')
```

```python
def criarRede():
    classificador = Sequential() # criando rede neural

    # fazendo mais camadas ocultas 
    classificador.add(Dense(units = 16, activation = 'relu', 
                            kernel_initializer = 'random_uniform', input_dim = 30)) # criando a primeira camada oculta 
    
    classificador.add(Dropout(0.2)) # camada de dropout (zera 20% do neuronios da camada de entrada)
    classificador.add(Dense(units = 16, activation = 'relu', 
                            kernel_initializer = 'random_uniform')) # segunda camada oculta 
    classificador.add(Dropout(0.2)) # camada de dropout
    
    classificador.add(Dense(units = 1, activation = 'sigmoid')) # camada de saida 

    otimizador = keras.optimizers.Adam(learning_rate = 0.001 , weight_decay = 0.0001, clipvalue = 0.5) # otimizador 

    classificador.compile(optimizer = otimizador, loss = 'binary_crossentropy', 
                        metrics = ['binary_accuracy']) # compila a rede neural  
    return classificador

# usando o scikeras
classificador = KerasClassifier(build_fn = criarRede, 
                                epochs = 100,
                                batch_size = 10)
```

```python
# obtendo resultados 
resultados = cross_val_score(estimator = classificador,
                             X = previsores, y = classe,
                             cv = 10, scoring = 'accuracy')
```

```python
print (resultados)
```

```python
media = resultados.mean() # calculando a media dos resultados
std = resultados.std() # calculando o desvio padrão (maior o valor maior a tendecia de ter overfitting)

print(f"Media: {media}\nDesvio Padrão: {std}")
```

### Tuning 

```python
import pandas as pd 
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV

# carregando os dados
previsores = pd.read_csv('entradas_breast.csv')
classe = pd.read_csv('saidas_breast.csv')
```

```python
# craiando a rede da mesma forma que foi criada anteriormente, porem agora com paramentros variaveis
def criarRede(optimizer, loos, kernel_initializer, activation, neurons):
    classificador = Sequential()
    classificador.add(Dense(units = neurons, activation = activation, 
                        kernel_initializer = kernel_initializer, input_dim = 30))
    classificador.add(Dropout(0.2))
    classificador.add(Dense(units = neurons, activation = activation, 
                        kernel_initializer = kernel_initializer))
    classificador.add(Dropout(0.2))
    classificador.add(Dense(units = 1, activation = 'sigmoid'))
    classificador.compile(optimizer = optimizer, loss = loos,
                      metrics = ['binary_accuracy'])
    return classificador

```

```python
classificador = KerasClassifier(build_fn = criarRede)
parametros = {'batch_size': [10, 30],
              'epochs': [50, 100],
              'optimizer': ['adam', 'sgd'],
              'loos': ['binary_crossentropy', 'hinge'],
              'kernel_initializer': ['random_uniform', 'normal'],
              'activation': ['relu', 'tanh'],
              'neurons': [16, 8]} # dicionario de parametros 
grid_search = GridSearchCV(estimator = classificador,
                           param_grid = parametros,
                           scoring = 'accuracy',
                           cv = 5)
# vai realizar os testes para poder descobrir qual os melhores parametros 
grid_search = grid_search.fit(previsores, classe)
melhores_parametros = grid_search.best_params_
melhor_precisao = grid_search.best_score_
```
