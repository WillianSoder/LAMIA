# Revisão de conceitos estatisticos

Este codigo visa aplicar os principais conceitos aprendidos no curso referente ao card 5 através de um exemplo basico de estatisticas de uma empresa e plotando um histograma das vendas mensais da empresa e um grafico de barras destas vendas 

```python
# importando as bibliotecas necessarias
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Criando os dados de vendas mensais em um ano 
vendas = [50000, 60000, 70000, 65000, 80000, 75000, 50000, 62000, 68000, 72000, 78000, 82000]
```

```python
# Calculando os conceitos estatísticos
media = np.mean(vendas)
mediana = np.median(vendas)
moda = stats.mode(vendas)  # Obtendo apenas o valor da moda
desvio_padrao = np.std(vendas)
```

```python
print(f"Média de vendas mensais: R${media}")
print(f"Mediana de vendas mensais: R${mediana}")
print(f"Moda de vendas mensais: R${moda[0]}") # para imprimir apenas o valor da moda
print(f"Desvio padrão de vendas mensais: {desvio_padrao}")
```

```python
# Plotando um histograma de vendas mensais
plt.figure()
plt.hist(vendas, bins=5, edgecolor='black')
plt.title('Distribuição de Vendas Mensais')
plt.xlabel('Vendas Mensais')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

```python
# Plotando um gráfico de barras das vendas mensais
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
plt.figure()
plt.bar(meses, vendas)
plt.title('Vendas Mensais por Mês')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
