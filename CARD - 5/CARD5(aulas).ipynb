Mean, Median, Mode em python usando numpy

```python
import numpy as np 

incomes = np.random.normal(27000, 15000, 10000)
# Calculando a media dos valores usando numpy 
np.mean(incomes)
```

```python
#para exibir o grafico dentro do notebook 
%matplotlib inline 
import matplotlib.pyplot as plt # importanto pyplot de matplolib para plotar o grafico
plt.hist(incomes, 50) # vai criar histograma dos dados (em 50 'buckets') 
plt.show() # imprime o garfico 
```

```python
# calculando a mediana 
np.median(incomes)
```

```python
# adicionando um ponto discrepante (valor grande) no conjunto de dados
incomes = np.append(incomes, [1000000000])

#calculando a mediana novamente
np.median(incomes) # o ponto discrepante não afetou a mediana 
```

```python
#calculando a media novamente 
np.mean(incomes) # é possivem observar que o ponto discrepante causou uma media errada 
```

```python
# Exponential PDF / "Power Law"
from scipy.stats import expon
x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

```python
# Função de massa binomial de probabilidade 
from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, 10, 0.001) # criando valores discretos 
plt.plot(x, binom.pmf(x, n, p)) # plotando o grafico 
```

```python
# função de masssa de probabilidade de poisson (parecido com a normal)
from scipy.stats import poisson

mu = 500
x = np.arange(400, 600 , 0.5) # criando conjunto de dados 
plt.plot(x, poisson.pmf(x, mu))
```

```python
# covariancia e correlação

# Função para centralizar os dados (de-mean) 
def de_mean(x):
    xmean = np.mean(x)
    return [xi - xmean for xi in x]


# Função para calcular a covariância
def covariance(x, y):
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n - 1)


# Gerar dados de exemplo
pageSpeeds = np.random.normal(3.0,1.0,1000)
purchaseAmount = np.random.normal(50.0,10.0,1000)

# Criar gráfico de dispersão
plt.scatter(pageSpeeds, purchaseAmount)

# Calcular e imprimir a covariância
cov = covariance(pageSpeeds, purchaseAmount)
print("Covariância:", cov)
```

```python
purchaseAmount = np.random.normal(50.0,10.0,1000) / pageSpeeds
# Criar gráfico de dispersão
plt.scatter(pageSpeeds, purchaseAmount)

# Calcular e imprimir a covariância
cov = covariance(pageSpeeds, purchaseAmount)
print("Covariância:", cov)
```

```python
# correlação 
def correlation(x,y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x,y)/ stddevx / stddevy

correlation(pageSpeeds, purchaseAmount)
```

```python
# calculando corelação com numpy 
np.corrcoef(pageSpeeds,purchaseAmount)
```

```python
# probabilidade condicional
from numpy import random
random.seed(0) # Definindo a semente do gerador de números aleatórios para garantir repetibilidade

# Dicionários para armazenar os totais de cada faixa etária e as compras
totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
totalPurchases = 0 # Variável para armazenar o total de compras

# Simulação de 100.000 eventos
for _ in range(100000):
    ageDecade = random.choice([20,30,40,50,60,70]) # Seleciona uma faixa etária aleatória
    purschaseProbability = float(ageDecade) / 100.0  # Calcula a probabilidade de compra com base na faixa etária
    totals[ageDecade] += 1  # Incrementa o total de ocorrências da faixa etária selecionada
    
     # Verifica se uma compra foi feita com base na probabilidade calculada
    if(random.random() < purschaseProbability):
        totalPurchases += 1
        purchases[ageDecade] += 1
```

```python
# exibindo resultados 
total = totals
comprados = purchases
total_comprados = totalPurchases
print(f"{total}\n\n{comprados}\n\n{total_comprados}")
```

```python
# probabilidade condicional (comprar algo na faixa de 30 anos)
PEF = float( purchases[30]) / float(totals[30])
print('P(purchase | 30s): ' + str(PEF))
```
