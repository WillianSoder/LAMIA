Mean, Median, Mode em python usando numpy

```python
import numpy as np 

incomes = np.random.normal(27000, 15000, 10000)
# Calculando a media dos valores usando numpy 
np.mean(incomes)
```

```python
#para exibir o grafico dentrp do notebook 
%matplotlib inline 
import matplotlib.pyplot as plt # importanto pyplot de matplolib para plotar o grafico
plt.hist(incomes, 50) # vai criar histograma dos dados (em 50 'buckets') 
plt.show() # imprime o garfico 
```

```python
# calculando a mediana 
np.median(incomes)
```

```python
# adicionando um ponto discrepante (valor grande) no conjunto de dados
incomes = np.append(incomes, [1000000000])

#calculando a mediana novamente
np.median(incomes) # o ponto discrepante não afetou a mediana 
```

```python
#calculando a media novamente 
np.mean(incomes) # é possivem observar que o ponto discrepante causou uma media errada 
```

```python
# MODA
ages = np.random.randint(18,high=90,size=500) # criando um conjunto de 500 idades entre 18 e 90 anos 
ages
```

```python
from scipy import stats # modulo de estatistica do scipy 
# Encontrando a moda
stats.mode(ages)
```

## Variancia e desvio padrão

```python
import numpy as np 
import matplotlib.pyplot as plt 

incomes = np.random.normal(100.0, 20.0, 10000) # crianda conjunto de dados

plt.hist(incomes, 50) # criando o histograma 
plt.show()
```

```python
# claculando o desvio padrão do conjunto de dados
incomes.std()
```

```python
# calculando a variancia do conunto de dados
incomes.var()
```

### exemplos de distribuição de dados

```python
# distribuição uniforme 
values = np.random.uniform(-10.0, 10.0, 100000) # criando um conjunto de dados com distribuição uniforme 
plt.hist(values, 50)
plt.show()
```

```python
# normal / gaussiana 
from scipy.stats import norm

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```

```python
mu = 5.0 # centralizador
sigma = 2.0 # desvio padrão 
values = np.random.normal(mu, sigma, 10000) # criando um conjunto de dados com distribuição uniforme 
plt.hist(values, 50)
plt.show()
```

```python
# Exponential PDF / "Power Law"
from scipy.stats import expon
x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

```python
# Função de massa binomial de probabilidade 
from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, 10, 0.001) # criando valores discretos 
plt.plot(x, binom.pmf(x, n, p)) # plotando o grafico 
```

```python
# função de masssa de probabilidade de poisson (parecido com a normal)
from scipy.stats import poisson

mu = 500
x = np.arange(400, 600 , 0.5) # criando conjunto de dados 
plt.plot(x, poisson.pmf(x, mu))
```

### Percentil

```python
values = np.random.normal(0, 0.5, 10000)# criando o conjunto de dados

plt.hist(values, 50) # criando histograma 
plt.show() # plotando o grafico 
```

```python
# calculando o percentil 50 (como se fosse a mediana) 
np.percentile(values, 50)
```

```python
# calculando o percentil 90 
np.percentile(values, 90)# 90% dos dados são inferiores ao ponto 
```

```python
# calculando o percentil 20
np.percentile(values, 20)# 20% dos dados são inferiores ao ponto 
```

moments: mean, variance, skew, kurtosis

```python
values = np.random.normal(0, 0.5, 10000) # criando conjunto de dados

plt.hist(values, 50) # criando histograma 
plt.show()
```

```python
# calculando media
np.mean(values)
```

```python
# calculando a variancia 
np.var(values)
```

```python
import scipy.stats as sp
# calculando a inclinação 
sp.skew(values)
```

```python
# calculando a curtose 
sp.kurtosis(values)
```
