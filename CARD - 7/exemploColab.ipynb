```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib
```

```python
# fazendo um grafico no notebook
t = np.arange(0.0, 2.0, 0.01)
s = 10 + np.exp(np.cos(2 * np.pi * t))

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='X', ylabel='Y',
       title='Colab example')
ax.grid()
```
