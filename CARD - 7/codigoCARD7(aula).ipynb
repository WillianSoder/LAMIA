```python
import numpy as np 
import matplotlib.pylab as plt 
import matplotlib
```

```python
# fazendo um grafico no notebook 
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Time (s)', ylabel='Voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()
```
