## aplicando conhecimentos obtidos nas aulas 


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# médias de notas dos alunos
notas_1_ano_EM = [7.5, 8.0, 6.5, 7.0, 8.5]
notas_2_ano_EM = [10.0, 9.5, 9.0, 4.0, 5.5]
notas_3_ano_EM = [7.3, 7.9, 8.8, 9.0, 9.5]
```

```python
# calculando medias das turmas usando numpy e exibindo
media_1 = np.mean(notas_1_ano_EM)
media_2 = np.mean(notas_2_ano_EM)
media_3 = np.mean(notas_3_ano_EM)

print(f"MEDIA DO 1º ANO: {media_1}\nMEDIA DO 2º ANO: {media_2}\nMEDIA DO 3º ANO: {media_3}\n")
```

```python
# criando os dados para o gráfico
anos = ['1º Ano', '2º Ano', '3º Ano']
medias = [media_1, media_2, media_3]
```

```python
# criando o gráfico
plt.bar(anos, medias, color=['gray', 'blue', 'red'])

# adicionando título e os rotulos dos eixos
plt.title('Médias de Notas dos Alunos')
plt.xlabel('Ano')
plt.ylabel('Média de Notas')
plt.show()
```
